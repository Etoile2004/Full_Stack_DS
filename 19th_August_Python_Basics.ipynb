Q1: Swap the values of two variables, x and y, without using a temporary variable in Python:

```python
# Assign integer values to x and y
x = 5
y = 10

# Swap the values without using a temporary variable
x = x + y
y = x - y
x = x - y

# Print the swapped values
print("x:", x)
print("y:", y)
```

Q2: Simple Python program that calculates the area of a rectangle by taking the length and width as inputs from the user

```python
# Get the length and width from the user
length = float(input("Enter the length of the rectangle: "))
width = float(input("Enter the width of the rectangle: "))

# Calculate the area
area = length * width

# Display the area
print("The area of the rectangle is:", area,"sqrd cm")
```

Q3:Python program that converts a temperature from Celsius to Fahrenheit:

```python
# Get the temperature in Celsius from the user
celsius = float(input("Enter the temperature in Celsius: "))

# Convert Celsius to Fahrenheit
fahrenheit = (celsius * 9/5) + 32

# Display the result
print("The temperature in Fahrenheit is:", fahrenheit,)
```

Q1: Python program that takes a string as input from the user and prints the length of the string

```python
## This program prompts the user to enter a string, calculates the length of the string
## using the len() function, and then displays the result.
```

```python
# Get the string input from the user
user_string = input("Enter a string: ")

# Calculate the length of the string
string_length = len(user_string)

# Display the length of the string
print("The length of the string is:", string_length)
```

Q2: Create a program that takes a sentence from the user and counts the number of vowels (a, e, i, o, u) in the string.

```python
# Get the sentence from the user
sentence = input("Enter a sentence: ")

# Initialize the count for vowels
vowel_count = 0

# Define the set of vowels
vowels = "aeiouAEIOU"

# Count the number of vowels in the sentence
for char in sentence:
    if char in vowels:
        vowel_count += 1

# Display the number of vowels
print("The number of vowels in the sentence is:", vowel_count)
```

Q3: Given a string, reverse the order of characters using string slicing and print the reversed string.

```python
# Get the string input from the user
user_string = input("Enter a string: ")

# Reverse the order of characters using string slicing
reversed_string = user_string[::-1]

# Print the reversed string
print("The reversed string is:", reversed_string)
```

Q4: Write a program that takes a string as input and checks if it is a palindrome (reads the same forwards and backwards).

```python
# Get the string input from the user
user_string = input("Enter a string: ")

# Remove any spaces and convert to lowercase for a case-insensitive comparison
normalized_string = user_string.replace(" ", "").lower()

# Check if the string is a palindrome
is_palindrome = normalized_string == normalized_string[::-1]

# Display the result
if is_palindrome:
    print("The string is a palindrome.")
else:
    print("The string is not a palindrome.")
```

Q5: Create a program that takes a string as input and removes all the spaces from it. Print the modified string without spaces.
